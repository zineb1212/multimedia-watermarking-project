# Watermarking Numérique par LSB

Ce notebook présente une implémentation pédagogique de la technique
LSB (Least Significant Bit) appliquée au watermarking d’images.

L’objectif est de :
- Comprendre le principe du LSB
- Insérer un watermark invisible dans une image
- Gérer les formats JPG et PNG comme dans une interface utilisateur réelle


```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

## Étape 2 : Chargement de l'image

```python
# Charger l'image en RGB
image_path = "data/image1.jpg"
image = Image.open(image_path).convert("RGB")
image_np = np.array(image)

print("Dimensions de l'image :", image_np.shape)
print("Type de données :", image_np.dtype)

# Affichage
plt.figure(figsize=(8, 6))
plt.imshow(image_np)
plt.title("Image originale")
plt.axis("off")
plt.show()
```

## Étape 3 : Préparation du message à cacher

```python
# Définir le message
watermark_text = "SECRET"

# Ajouter un délimiteur pour marquer la fin du message
delimiter = "###END###"
message_complet = watermark_text + delimiter

# Convertir le message en chaîne binaire
message_binaire = ''.join(format(ord(char), '08b') for char in message_complet)

print("Message à cacher :", watermark_text)
print("Longueur du message (caractères) :", len(watermark_text))
print("Longueur en bits :", len(message_binaire))
print("Exemple de conversion :", watermark_text[0], "->", format(ord(watermark_text[0]), '08b'))
```

## Étape 4 : Vérification de la capacité de l'image

```python
# Calculer la capacité de l'image
capacite_image = image_np.size
bits_necessaires = len(message_binaire)

print("Capacité de l'image (en bits) :", capacite_image)
print("Bits nécessaires pour le message :", bits_necessaires)
print("Taux d'occupation :", f"{(bits_necessaires / capacite_image) * 100:.2f}%")

if bits_necessaires > capacite_image:
    print("ERREUR : Le message est trop long pour cette image")
else:
    print("OK : L'image peut contenir ce message")
```

## Étape 5 : Insertion du watermark (méthode LSB avec support multi-bits)

Cette étape permet de choisir le nombre de bits LSB à modifier (1, 2, 3 ou 4 bits).
- 1 bit : Invisible, qualité maximale
- 2 bits : Subtile, bon compromis
- 3 bits : Robuste, légèrement visible
- 4 bits : Très robuste, visible

```python
# Paramètre : nombre de bits LSB à modifier (1, 2, 3 ou 4)
lsb_bits = 4  # Changez cette valeur pour tester 1, 2, 3 ou 4 bits

# Copier l'image pour préserver l'originale
image_watermarkee = image_np.copy()

# Aplatir l'image en un vecteur 1D
pixels_plats = image_watermarkee.flatten()

# Créer le masque pour effacer les bits LSB
# Pour 1 bit: 0xFE (11111110), pour 2 bits: 0xFC (11111100), etc.
mask = ~((1 << lsb_bits) - 1) & 0xFF

print(f"Utilisation de {lsb_bits} bit(s) LSB")
print(f"Masque utilisé: 0x{mask:02X} (binaire: {bin(mask)[2:].zfill(8)})")

# Insérer le message
bit_index = 0
pixels_modifies = 0

for i in range(len(pixels_plats)):
    if bit_index >= len(message_binaire):
        break
    
    # Effacer les bits LSB
    pixels_plats[i] = pixels_plats[i] & mask
    
    # Insérer les bits du message (1 à lsb_bits bits par pixel)
    bits_to_insert = 0
    for j in range(lsb_bits):
        if bit_index < len(message_binaire):
            bits_to_insert = (bits_to_insert << 1) | int(message_binaire[bit_index])
            bit_index += 1
    
    # Insérer les bits dans le pixel
    pixels_plats[i] = pixels_plats[i] | bits_to_insert
    
    if bits_to_insert > 0:
        pixels_modifies += 1

# Reconstruire l'image
image_watermarkee = pixels_plats.reshape(image_np.shape)

print("Insertion terminée")
print(f"Nombre de pixels modifiés : {pixels_modifies}")
print(f"Bits insérés : {bit_index} / {len(message_binaire)}")
```

## Étape 6 : Comparaison visuelle

```python
# Affichage côte à côte
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].imshow(image_np)
axes[0].set_title("Image originale")
axes[0].axis("off")

axes[1].imshow(image_watermarkee)
axes[1].set_title("Image avec watermark LSB")
axes[1].axis("off")

plt.tight_layout()
plt.show()

print("Observation : Les deux images semblent identiques à l'oeil nu")
```

## Étape 7 : Sauvegarde de l'image watermarkée

```python
# Sauvegarder en PNG pour préserver les bits
output_path = "image_watermarkee.png"
Image.fromarray(image_watermarkee.astype('uint8')).save(output_path)
print("Image sauvegardée :", output_path)
```

## Étape 8 : Extraction du watermark (avec support multi-bits)

Important : Utilisez le même nombre de bits LSB que lors de l'insertion !

```python
# IMPORTANT : Utiliser le même nombre de bits que lors de l'insertion
lsb_bits_extraction = 4  # Doit correspondre à lsb_bits utilisé lors de l'insertion

# Charger l'image watermarkée
image_chargee = np.array(Image.open(output_path))
pixels_plats = image_chargee.flatten()

# Créer le masque pour extraire les bits LSB
mask_extraction = (1 << lsb_bits_extraction) - 1

# Extraire les bits LSB
bits_extraits = ''
for pixel in pixels_plats:
    lsb_value = pixel & mask_extraction
    # Convertir la valeur en binaire avec le bon nombre de bits
    bits = format(lsb_value, f'0{lsb_bits_extraction}b')
    bits_extraits += bits

print(f"Extraction avec {lsb_bits_extraction} bit(s) LSB")
print(f"Nombre total de bits extraits : {len(bits_extraits)}")

# Convertir les bits en caractères
caracteres = []
for i in range(0, len(bits_extraits), 8):
    byte = bits_extraits[i:i+8]
    if len(byte) == 8:
        caractere = chr(int(byte, 2))
        caracteres.append(caractere)
        
        # Arrêter au délimiteur
        texte_actuel = ''.join(caracteres)
        if delimiter in texte_actuel:
            break

# Extraire le message sans le délimiteur
message_extrait = ''.join(caracteres).split(delimiter)[0]
print("\nMessage extrait :", message_extrait)
print("Message original :", watermark_text)
print("Extraction réussie :", message_extrait == watermark_text)
```

## Étape 9 : Évaluation de la qualité (PSNR et MSE)

Le PSNR (Peak Signal-to-Noise Ratio) et le MSE (Mean Squared Error) permettent de quantifier la différence entre l'image originale et l'image watermarkée.

Interprétation du PSNR :
- PSNR > 40 dB : Excellente qualité, différence imperceptible
- PSNR entre 30 et 40 dB : Bonne qualité
- PSNR < 30 dB : Qualité dégradée, différence visible

```python
# Calcul du MSE
mse = np.mean((image_np.astype(float) - image_watermarkee.astype(float)) ** 2)

# Calcul du PSNR
if mse == 0:
    psnr = float('inf')
    print("MSE : 0 (images identiques)")
    print("PSNR : infini")
else:
    psnr = 20 * np.log10(255.0 / np.sqrt(mse))
    print(f"MSE : {mse:.6f}")
    print(f"PSNR : {psnr:.2f} dB")
    
    if psnr > 40:
        print("Qualité : Excellente (imperceptible)")
    elif psnr > 30:
        print("Qualité : Bonne")
    else:
        print("Qualité : Dégradée")
```

## Étape 10 : Analyse des différences pixel par pixel

```python
# Calculer la différence absolue entre les deux images
difference = np.abs(image_np.astype(int) - image_watermarkee.astype(int))

print("Statistiques des différences :")
print(f"  Différence moyenne : {difference.mean():.4f}")
print(f"  Différence maximale : {difference.max()}")
print(f"  Différence minimale : {difference.min()}")
print(f"  Nombre de pixels modifiés : {np.count_nonzero(difference)}")
print(f"  Pourcentage de pixels modifiés : {(np.count_nonzero(difference) / image_np.size) * 100:.2f}%")

# Visualisation de la différence (amplifiée)
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image_np)
plt.title("Image originale")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(image_watermarkee)
plt.title("Image watermarkée")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(difference * 50, cmap='hot')
plt.title("Différence (amplifiée x50)")
plt.colorbar(label="Intensité")
plt.axis("off")

plt.tight_layout()
plt.show()
```
